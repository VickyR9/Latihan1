```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["TI", "SI", "DKV", "TS"])
y = np.array([90, 80, 65, 79])

plt.xlabel('Departmen')
plt.ylabel('Number of student')

plt.title('Total Student')

plt.bar(x,y)
plt.show()
```
